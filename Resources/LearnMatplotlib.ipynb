```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Type of Plots

### Line Plots

```python
x = [1,2,3,4,5,6]
y = [10,25,30,49,50,60]
```

```python
plt.plot(x,y);
plt.title("Simple Line Plot")
plt.xlabel('X-axis Label');
plt.ylabel('Y-axis Label');
plt.grid(True)
```

### Bar Chart

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
revenue = [1100,232,654,1212]
```

```python
plt.figure(figsize=(10,10))
plt.bar(months, revenue)
plt.title("Monthly Revenue Bar Chart")
plt.xlabel('Months');
plt.ylabel('Revenue');
```

### Subplots & Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 15, 20, 25,30]
predicted_price = [8, 18, 15, 29,50]
```

```python
plt.plot(days, original_price, 'b-o',label = "Original Price");
plt.plot(days, predicted_price, 'r-o',label = "Predicted Price");
plt.legend()
plt.title("Plot with Legends")
plt.xlabel("Weekdays")
plt.ylabel("Price")
```

### Subplots

```python
# First subplot - Original Price
plt.subplot(3,1,1)
plt.plot(days, original_price, 'b-o',label = 'Original Price')
plt.title('Original Prices over Time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

# Second subplot - Predicted Price
plt.subplot(3,1,2)
plt.plot(days, original_price, 'r-o',label = 'Predicted Price')
plt.title('Predicted Prices over Time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

# Third subplot - Difference between Original & Predicted Price
price_difference = [ o-p for o, p in zip(original_price, predicted_price)]
plt.subplot(3,1,3)
plt.bar(days, price_difference, color = 'purple',label = 'Price Difference')
plt.title('Difference between Original & Predicted Price')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

plt.tight_layout()
```


### Saving the Figure

```python
plt.plot(days, original_price, 'b-o',label = "Original Price");
plt.plot(days, predicted_price, 'r-o',label = "Predicted Price");
plt.legend()
plt.title("Plot with Legends")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.savefig('example_prediction.png', dpi = 300) #save the image
```
